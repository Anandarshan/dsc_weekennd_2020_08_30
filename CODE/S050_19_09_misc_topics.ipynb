```python
def rec_sum(x):
    if x == 0: # terminating condition for recursion
        return 0
    
    return x + rec_sum(x-1)


print(rec_sum(5))

# rec_sum(5)
#           5 + rec_sum(4)
#                          4 + rec_sum(3)
#                                        3 + rec_sum(2)
#                                                       2 + rec_sum(1)
#                                                                     1 + rec_sum(0)
#                                                                                   retunr 0
```

```python
# nested loops
ch = 65
for i in range(5): # outer

    for j in range(5-i): # inner: space
        print(" ", end='')
    for j in range(i+1): # inner: symbol
        print(chr(ch), end='')
        ch += 1
    print()
```

```python
# enumerate : find position of 30 in this list
a = [10,20,30,40,50]

idx = -1
for i in range(len(a)):
    if a[i] == 30:
        idx = i
        break
        
print(idx)

print()
idx = -1
for i, num in enumerate(a):
    if num == 30:
        idx = i
        break
        
print(idx)
```

```python
# variable length arguments
def f(*args):
    print("Args:", args, len(args))
    
f()
f(10)
f(10,20,33)
```
